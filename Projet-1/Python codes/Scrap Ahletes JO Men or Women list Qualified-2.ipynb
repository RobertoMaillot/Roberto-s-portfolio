```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

# URL à scraper
url = "https://worldathletics.org/stats-zone/road-to/7153115?eventId=10229534"

# Configuration du navigateur
driver = webdriver.Chrome()

# Envoi de la requête et récupération du contenu HTML
driver.get(url)

# Trouver les éléments td directement
td_elements = driver.find_elements(By.CSS_SELECTOR, "tr > td")

# Diviser les données en groupes de six (si chaque ligne a six colonnes)
data = [td.text.strip() for td in td_elements]
grouped_data = [data[i:i+6] for i in range(0, len(data), 6)]

# Colonnes du DataFrame
columns = ['QP', 'CP', 'Nat', 'Name', 'Statut', 'Details']

# Créer le DataFrame
df = pd.DataFrame(grouped_data, columns=columns)

# Afficher le DataFrame
print(df)

# Fermer le navigateur
driver.quit()
```
