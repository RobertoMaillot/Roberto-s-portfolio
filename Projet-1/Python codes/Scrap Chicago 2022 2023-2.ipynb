```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Initialize an empty DataFrame to store the final results
df_final_men_2023 = pd.DataFrame()

for i in range(1, 27):
    # URL to scrape
    #menurl = f"https://results.chicagomarathon.com/2023/?page{i}&event=MAR&num_results=1000&pid=search&pidp=start&search%5Bage_class%5D=%25&search%5Bsex%5D=M&search%5Bnation%5D=%25&search%5Bstate%5D=%25&search_sort=place_nosex"
    #url = f"https://results.chicagomarathon.com/2023/?page={i}&event=MAR&num_results=1000&pid=search&pidp=start&search%5Bage_class%5D=%25&search%5Bsex%5D=W&search%5Bnation%5D=%25&search%5Bstate%5D=%25&search_sort=place_nosex"
    url = f"https://results.chicagomarathon.com/2023/?page={i}&event=MAR&num_results=1000&pid=list&pidp=start&search%5Bsex%5D=M&search%5Bage_class%5D=%25"
    #url = f"https://results.chicagomarathon.com/2023/?page={i}&event=MAR&num_results=1000&pid=list&pidp=start&search%5Bsex%5D=W&search%5Bage_class%5D=%25"
    # Browser configuration
    driver = webdriver.Chrome()
    driver.get(url)

    # Allow some time for the page to load
    time.sleep(2)

    # Extract text content from all elements with class "list-group-item row"
    elements = driver.find_elements(By.CLASS_NAME, "list-group-item.row")
    data_list = [element.text for element in elements]

    # Convert the list into a DataFrame
    df1 = pd.DataFrame(data_list, columns=["Data"])

    # Concatenate df1 to df_final
    df_final_men_2023 = pd.concat([df_final_men_2023, df1], ignore_index=True)

    # Close the browser when done
    driver.quit()

# Display the final DataFrame
df_final_men_2023
```
