```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
ranks = []
names = []
countries = []
start_dates = []
end_dates = []
part_scores = []
result_scores = []
comp_scores = []
urls = []
```

```python
response= requests.get('https://worldathletics.org/records/competition-performance-rankings?type=7&year=2022&sortBy=score&page=2')
```

```python
page_contents = response.text
```

```python
soup = bs(response.text, 'lxml')
```

```python
import re 

def Rank(soup): # Gets the rank of the competition
    td_tags = soup.find_all('td', {'data-th' : "Place"})
    for tag in td_tags:
        ranks.append(tag.text.strip())
    return ranks

def meeting_name(soup): # Gets the edition of the competition
    td_tags = soup.find_all('td', {'data-th' : "Competition"})
    for tag in td_tags:
        names.append(tag.text.strip())      
    return names

def country(soup): # Gets the country the competition is held
    country= soup.find_all('td',{'data-th' : "Country"} )
    for tags in country:
        countries.append(tags.text.strip())
    return countries

def Start_Date(soup): # Gets the date the competition is held
    start_date = soup.find_all('td', {'data-th' : "Start Date"})
    for tags in start_date:
        start_dates.append(tags.text.strip())
    return start_dates

def End_Date(soup): # Gets the date the competition is held
    end_date = soup.find_all('td', {'data-th' : "End Date"})
    for tags in end_date:
        end_dates.append(tags.text.strip())
    return end_dates

def Part_Score(soup): # Gets the date the competition is held
    part_score = soup.find_all('td', {'data-th' : "Part. Score"})
    for tags in part_score:
        part_scores.append(tags.text.strip())
    return part_scores

def Result_Score(soup): # Gets the date the competition is held
    result_score = soup.find_all('td', {'data-th' : "Result Score"})
    for tags in result_score:
        result_scores.append(tags.text.strip())
    return result_scores

def Comp_Score(soup): # Gets the date the competition is held
    comp_score = soup.find_all('td', {'data-th' : "Comp. Score"})
    for tags in comp_score:
        comp_scores.append(tags.text.strip())
    return comp_scores

def Event_url(soup): # Gets the links to the results
    tbody = soup.find('tbody')
    tags = tbody.find_all('tr')
        ### the`re.compile gets href with the given string`
    base_url = 'https://www.worldathletics.org'
    for tag in tags:
        urls.append(base_url + tag.attrs.get('data-athlete-url', 'Not found'))
    return urls
```

```python
athletics_dict = {
'Rank': Rank(soup),
'meeting_name': meeting_name(soup),
'country': country(soup),
'Start Date': Start_Date(soup),
'End Date': End_Date(soup),
'Participation Score': Part_Score(soup),
'Result Score': Result_Score(soup),
'Competition Score': Comp_Score(soup),
'Event URL': Event_url(soup)
}        
```

```python
df = pd.DataFrame(athletics_dict)
df
```
