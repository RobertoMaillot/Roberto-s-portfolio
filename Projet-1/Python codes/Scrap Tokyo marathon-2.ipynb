```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup

# URL à scraper
url = "https://www.marathon.tokyo/2024/result/index.php"

# Configuration du navigateur
driver = webdriver.Chrome()

driver.get(url)

time.sleep(2)

access_button = driver.find_element(By.CSS_SELECTOR, "#btn_submit")
access_button.click()

time.sleep(2)

# Initial scrape for the first page
soup = BeautifulSoup(driver.page_source, 'lxml')
df1 = pd.read_html(soup.select_one('.m-item_tbl').prettify())[0]

# Loop through pages 2 to 715
for i in range(1, 715):
    # Wait for a short time to ensure the page is loaded
    time.sleep(2)

    # Get the current page's HTML
    soup = BeautifulSoup(driver.page_source, 'lxml')

    # Read the table data into a DataFrame
    df = pd.read_html(soup.select_one('.m-item_tbl').prettify())[0]

    # Assuming you want to concatenate the dataframes, adjust this according to your needs
    df1 = pd.concat([df1, df], ignore_index=True)

    # Click on the next page button
    pagination_button = driver.execute_script(f"page({i + 1});")

# Close the browser when done
driver.quit()

# Now df1 contains the data from all pages
```
