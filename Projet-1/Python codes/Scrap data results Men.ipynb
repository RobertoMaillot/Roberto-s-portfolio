```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
ranks = []
Marks = []
WINDS = []
Competitors = []
DOBS = []
Nats = []
Poss = []
Venues = []
dates = []
ResultScores = []
urls = []
```

```python
def extract_data(soup):
    # Extract data using BeautifulSoup
    td_tags = soup.find_all('td', {'data-th': 'Rank'})
    for tag in td_tags:
        ranks.append(tag.text.strip())

    Mark = soup.find_all('td', {'data-th': "Mark"})
    for tag in Mark:
        Marks.append(tag.text.strip())

    WIND = soup.find_all('td', {'data-th': "WIND"})
    for tag in WIND:
        WINDS.append(tag.text.strip())

    Competitor = soup.find_all('td', {'data-th': "Competitor"})
    for tag in Competitor:
        Competitors.append(tag.text.strip())

    DOB = soup.find_all('td', {'data-th': "DOB"})
    for tag in DOB:
        DOBS.append(tag.text.strip())

    Nat = soup.find_all('td', {'data-th': "Nat"})
    for tag in Nat:
        Nats.append(tag.text.strip())

    Pos = soup.find_all('td', {'data-th': "Pos"})
    for tag in Pos:
        Poss.append(tag.text.strip())

    Venue = soup.find_all('td', {'data-th': "Venue"})
    for tag in Venue:
        Venues.append(tag.text.strip())

    date = soup.find_all('td', {'data-th': "Date"})
    for tag in date:
        dates.append(tag.text.strip())

    ResultScore = soup.find_all('td', {'data-th': "ResultScore"})
    for tag in ResultScore:
        ResultScores.append(tag.text.strip())

    # Extract URLs
    tbody = soup.find('tbody')
    tags = tbody.find_all('a')
    base_url = 'https://www.worldathletics.org/'
    for tag in tags:
        urls.append(base_url + tag.attrs.get('href', 'Not found'))

# Loop through pages
for i in range(1, 7):
    response = requests.get(f'https://worldathletics.org/records/toplists/road-running/marathon/all/men/senior/2001?regionType=world&page={i}&bestResultsOnly=false&maxResultsByCountry=all&eventId=10229634&ageCategory=senior')
    soup = bs(response.text, 'lxml')

    # Call the function to extract data
    extract_data(soup)

# Now you have all the data in the lists (names, Marks, WINDS, etc.)
# You can use these lists to create a DataFrame or perform further analysis.
```

```python
athletics_dict = {
    'Rank': ranks,
    'Mark':Marks,
    'WIND':WINDS,
    'Competitor':Competitors,
    'DOB':DOBS,
    'Nat':Nats,
    'Pos':Poss,
    'Venue':Venues,
    'Date':dates,
    'ResultScore':ResultScores,
    'athletes_url':urls
    }
```

```python
pd.DataFrame(athletics_dict)
```

```python
df = pd.DataFrame(athletics_dict)
```

```python
df.to_csv('results_men_2001.csv', index=False)
```

```python
import pandas as pd

# Liste des années
years = list(range(2001, 2025))

# Initialisez un DataFrame vide
union_df = pd.DataFrame()

# Bouclez à travers les années et effectuez une union join
for year in years:
    file_path = f"results_men_{year}.csv"
    
    try:
        # Chargez le fichier CSV correspondant à l'année
        df = pd.read_csv(file_path)
        
        # Ajoutez une colonne avec l'année
        df['Year'] = year
        
        # Effectuez une union join avec le DataFrame principal
        union_df = pd.concat([union_df, df], ignore_index=True)
        
        print(f"File {file_path} successfully loaded and union joined.")
    except FileNotFoundError:
        print(f"File {file_path} not found.")

# Affichez le DataFrame avec l'union join
print(union_df)

# Sauvegardez le DataFrame avec l'union join dans un nouveau fichier CSV
union_df.to_csv('union_joined_results_men.csv', index=False)
```
