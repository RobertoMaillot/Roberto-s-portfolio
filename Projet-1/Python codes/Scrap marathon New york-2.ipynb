```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

for i in range(1,51350,50) :
    # Open the url in the webdriver
    #url = f'https://results.nyrr.org/event/M2023/finishers?_gl=1*e10me0*_gcl_au*MTUwMzg2MDI4OS4xNzA5NjM4MzU1#page={i}'
    url = f'https://results.nyrr.org/event/M2023/finishers?_gl=1*e10me0*_gcl_au*MTUwMzg2MDI4OS4xNzA5NjM4MzU1#page=1&opf=1'
    driver = webdriver.Chrome()
    driver.get(url)

    # Wait for the page to load
    time.sleep(3)

    # Find all elements with class name 'b-runners_detailes-list'
    lists = driver.find_elements(By.CLASS_NAME, "rms-grid-item")

    # Initialize an empty list to store data
    data = []

    # Extract information from each 'l' element
    for l in lists:
        lines = l.text.split('\n')

        # Check if the line has the expected number of values
        if len(lines) >= 9:
            # Extracting information from the lines
            name = lines[0]
            age_category = lines[1]
            country_bib = lines[2]
            info_time = lines[4]
            pace = lines[6]
            place = lines[8]

            # Extracting specific details from the 'age_category' string
            age, category = age_category.split()[:2]

            # Extracting specific details from the 'country_bib' string
            country_bib_parts = country_bib.split()
            if len(country_bib_parts) >= 3:
                country, bib = country_bib_parts[1:3]
            else:
                country, bib = None, None

            # Append the extracted data to the list
            data.append({
                'Nom': name,
                'Catégorie&Âge': age_category,
                #'Country': country,
                #'Bib': bib,
                'Place': place,
                'InfoTime': info_time,
                'Pace': pace
            })
            
            #Quit the driver
            driver.implicitly_wait(100)
            driver.quit()
            
            
            url = f'https://results.nyrr.org/event/M2023/finishers?_gl=1*e10me0*_gcl_au*MTUwMzg2MDI4OS4xNzA5NjM4MzU1#page=1&opf={i}'
            driver = webdriver.Chrome()
            driver.get(url)
            

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```
